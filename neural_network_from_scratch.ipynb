```python
import math
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('/content/sample_data/mnist_train_small.csv')
```

```python
df
```

```python
learning_rate=0.01
```

```python
def sigmoid(z):
  return (1/(1+np.exp(-z)))
```

```python
w1 = np.random.uniform(-0.5, 0.5, (30, 784))
b1 = np.zeros((30, 1))
```

```python
w2= np.random.uniform(-0.5,0.5,(10,30))
b2=np.zeros((10,1))
```

```python
epochs=5
```

```python
for __ in range(epochs):
  count=0
  for i in range (19999):
    first_row = df.iloc[i]


    y=np.zeros((10, 1))
    y[first_row[0]]=1

    x1= (first_row.drop(first_row.index[0]))/255 #x is the value input
    x_1=x1.values
    x=x_1.reshape((784, 1))

    z1=w1@x + b1
    h1=sigmoid(z1)

    z2=w2@h1 + b2
    h2=sigmoid(z2)

    if np.argmax(h2) == np.argmax(y):
      count=count+1

    delta=h2-y
    w2+=-learning_rate*delta @ np.transpose(h1)
    b2+= -learning_rate * delta

    del_h=np.transpose(w2) @ delta * (h1 * (1-h1))
    w1 += -learning_rate * del_h @ np.transpose(x)
    b1 += -learning_rate * del_h

  accuracy= (count/19999)*100
  print(accuracy)
```
